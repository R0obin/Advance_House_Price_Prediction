```python
# Lets handle the null values
# first start with categorical values
feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()> 1 and dataset[feature].dtypes=='O']
for feature in feature_nan:
  print(feature, np.round(dataset[feature].isnull().mean(),4),'% of missing values')
```

```python
def replace_null_values(dataset,feature_nan):
  data=dataset.copy()
  data[feature_nan]=data[feature_nan].fillna('Missing')
  return data
dataset= replace_null_values(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
```

```python
for feature in feature_nan:
  print(feature,dataset[feature].unique())
```

```python
## Handling the Numeric Variable with null values
numeric_feature = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtypes !='O']
print(len(numeric_feature))
numeric_feature
for feature in numeric_feature:
  print(feature,np.round(dataset[feature].isnull().mean(),4), "% of missing values ")
```

```python
dataset.head(50)
```

```python
# # Replacing the nan values with median
for feature in numeric_feature:
  median_value = dataset[feature].median()
  dataset[feature + 'nan']= np.where(dataset[feature].isnull(), 1, 0)
  dataset[feature].fillna(median_value,inplace=True)
dataset[numeric_feature].isnull().sum()
```

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in num_features:
  dataset[feature]= np.log(dataset[feature])
dataset[num_features].head()
```

```python
## Handling the Rare Categorical
# we will remove the cateforical variables whch have less than 1 % value form the observation
Categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O' ]
Categorical_feature
```

**Finding the Rare Categorical Varivable**
# if categorical variable values is less than 1% changing label into Rare_var

```python
for feature in Categorical_feature:
  temp = dataset.groupby(feature)['SalePrice'].count() / len(dataset)
  temp_df = temp[temp > 0.01].index
  dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature], 'Rare_var')
dataset[Categorical_feature].head(50)
```

```python
for feature in Categorical_feature:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```
