```python
# Capture the dependent variable
y_train = dataset['SalePrice']
```

```python
# droping the dependent variable form the entire dataset
X_train = dataset.drop(['Id','SalePrice'],axis=1)
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
# let's print the number of selected features
selected_feature = X_train.columns[(feature_sel_model.get_support())]

#Lets see the some calculation of rest features

print('Total number of features: {}'.format(np.sum(X_train.shape[1])))
print('Selected features from model: {}'.format(len(selected_feature)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_==0)))
```
